```python
import pandas as pd
from collections import Counter

# 读取CSV文件
def read_csv(file_path):
    return pd.read_csv(file_path)

# 统计词频
def count_word_frequency(data_series):
    word_counts = Counter()
    # 遍历每一行数据
    for item in data_series.dropna():
        # 将文本拆分成短语，同时考虑中文逗号和英文逗号
        phrases = item.replace('，', ',').replace(',', '、').split('、')
        # 遍历每个短语，进一步拆分成词语
        for phrase in phrases:
            words = phrase.split()
            # 更新词频统计
            word_counts.update(word for word in words if word)  # 确保不统计空字符串
    
    return word_counts

# 主函数
def main():
    # CSV文件路径
    file_path = '动机分析_苏泊尔_自由_1028_version6.csv'  # 确保这是正确的文件路径
    # 读取CSV文件
    df = read_csv(file_path)
    # 统计“AI分析”列的词频
    word_counts = count_word_frequency(df['AI分析'])
    # 输出词频结果
    for word, count in word_counts.items():
        print(f"{word}: {count}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from collections import Counter

# 读取CSV文件
def read_csv(file_path):
    return pd.read_csv(file_path)

# 统计词频
def count_word_frequency(data_series):
    word_counts = Counter()
    # 遍历每一行数据
    for item in data_series.dropna():
        # 将文本拆分成短语，同时考虑中文逗号和英文逗号
        phrases = item.replace('，', ',').replace(',', '、').split('、')
        # 遍历每个短语，进一步拆分成词语
        for phrase in phrases:
            words = phrase.split()
            # 更新词频统计
            word_counts.update(word for word in words if word)  # 确保不统计空字符串
    
    return word_counts

# 保存词频结果到Excel文件
def save_to_excel(word_counts, output_file):
    # 将Counter对象转换为DataFrame
    df_word_counts = pd.DataFrame(word_counts.items(), columns=['Word', 'Frequency'])
    # 将DataFrame保存到Excel文件
    df_word_counts.to_excel(output_file, index=False)

# 主函数
def main():
    # CSV文件路径
    file_path = '动机分析_苏泊尔_自由_1028_version6.csv'  # 确保这是正确的文件路径
    # Excel文件输出路径
    output_file = 'word_frequency.xlsx'  # 你可以根据需要修改文件名和路径
    # 读取CSV文件
    df = read_csv(file_path)
    # 统计“AI分析”列的词频
    word_counts = count_word_frequency(df['AI分析'])
    # 保存词频结果到Excel文件
    save_to_excel(word_counts, output_file)
    print(f"词频统计结果已保存到'{output_file}'")

if __name__ == "__main__":
    main()
```
